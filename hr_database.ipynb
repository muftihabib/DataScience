```python
import pymysql
import pandas as pd
connection = pymysql.connect(
    host='localhost',
    user='root',
    password= 'tahseen',
    database='hr'
)
cursor= connection.cursor()
```

```python
# 1. Retrieve all details of employee
select_query1 = """SELECT * FROM employees"""
cursor.execute(select_query1)
rows1 = cursor.fetchall()
df1 = pd.DataFrame(rows1, columns=[col[0] for col in cursor.description])
display(df1)
```

```python
# 2. Display first name, last name, and email of all employees
select_query2 = """SELECT first_name, last_name, email FROM employees"""
cursor.execute(select_query2)
rows2 = cursor.fetchall()
df2 = pd.DataFrame(rows2, columns=['first_name', 'last_name', 'email'])
display(df2)
```

```python
# 3. Retrieve distinct job titles from the jobs table
select_query3 = """SELECT DISTINCT job_title FROM jobs"""
cursor.execute(select_query3)
rows3 = cursor.fetchall()
df3 = pd.DataFrame(rows3, columns=['job_title'])
display(df3)
```

```python
# 4. Find the total number of employees in the company
select_query4 = """SELECT COUNT(*) AS total_employees FROM employees"""
cursor.execute(select_query4)
rows4 = cursor.fetchall()
df4 = pd.DataFrame(rows4, columns=['total_employees'])
display(df4)
```

```python
# 5. Retrieve employees hired after January 1, 2015
select_query5 = """SELECT * FROM employees WHERE hire_date > '2015-01-01'"""
cursor.execute(select_query5)
rows5 = cursor.fetchall()
df5 = pd.DataFrame(rows5, columns=[col[0] for col in cursor.description])
display(df5)
```

```python
# 6. List all employees who have a salary greater than 5000
query6 = """SELECT * FROM employees WHERE salary > 5000"""
cursor.execute(query6)
rows6 = cursor.fetchall()
df6 = pd.DataFrame(rows6, columns=[col[0] for col in cursor.description])
display(df6)
```

```python
query7 = """SELECT e.* FROM employees e JOIN jobs j ON e.job_id = j.job_id WHERE j.job_title LIKE '%Manager%'"""
cursor.execute(query7)
rows7 = cursor.fetchall()
df7 = pd.DataFrame(rows7, columns=[col[0] for col in cursor.description])
display(df7)
```

```python
# 8. Retrieve all employees whose first name starts with ‘A’ and ends with ‘n’
query8 = """SELECT * FROM employees WHERE first_name LIKE 'A%n'"""
cursor.execute(query8)
rows8 = cursor.fetchall()
df8 = pd.DataFrame(rows8, columns=[col[0] for col in cursor.description])
display(df8)
```

```python

# 9. Display the employees who do not have a commission
query9 = """SELECT * FROM employees WHERE commission_pct IS NULL OR commission_pct = 0"""
cursor.execute(query9)
rows9 = cursor.fetchall()
df9 = pd.DataFrame(rows9, columns=[col[0] for col in cursor.description])
display(df9)
```

```python
# 10. Retrieve the top 5 highest-paid employees
query10 = """SELECT * FROM employees ORDER BY salary DESC LIMIT 5"""
cursor.execute(query10)
rows10 = cursor.fetchall()
df10 = pd.DataFrame(rows10, columns=[col[0] for col in cursor.description])
display(df10)
```

SQL FUNCTIONS

```python
# 11. Find the average salary of all employees
query11 = """SELECT AVG(salary) AS average_salary FROM employees"""
cursor.execute(query11)
rows11 = cursor.fetchall()
df11 = pd.DataFrame(rows11, columns=[col[0] for col in cursor.description])
display(df11)
```

```python
# 12. Retrieve the total number of employees working in each department
query12 = """SELECT department_id, COUNT(*) AS total_employees FROM employees GROUP BY department_id"""
cursor.execute(query12)
rows12 = cursor.fetchall()
df12 = pd.DataFrame(rows12, columns=[col[0] for col in cursor.description])
display(df12)
```

```python
# 13. Display the employee's first name and the length of their first name
query13 = """SELECT first_name, LENGTH(first_name) AS name_length FROM employees"""
cursor.execute(query13)
rows13 = cursor.fetchall()
df13 = pd.DataFrame(rows13, columns=[col[0] for col in cursor.description])
display(df13)
```

```python
# 14. Convert the hire_date of employees to display only the year
query14 = """SELECT first_name, YEAR(hire_date) AS hire_year FROM employees"""
cursor.execute(query14)
rows14 = cursor.fetchall()
df14 = pd.DataFrame(rows14, columns=[col[0] for col in cursor.description])
display(df14)
```

```python
query15 = """SELECT j.job_title, MIN(e.salary) AS min_salary, MAX(e.salary) AS max_salary 
             FROM employees e JOIN jobs j ON e.job_id = j.job_id 
             GROUP BY j.job_title"""
cursor.execute(query15)
rows15 = cursor.fetchall()
df15 = pd.DataFrame(rows15, columns=[col[0] for col in cursor.description])
display(df15)
```

✔ AVG() for average salary
✔ COUNT() + GROUP BY for department-wise employee count
✔ LENGTH() to calculate the length of first names
✔ YEAR() function to extract the year from hire_date 
✔ MIN() & MAX() + GROUP BY for job-wise salary range 

```python
query16 = """
SELECT e.first_name, e.last_name, d.department_name
FROM employees e
JOIN departments d ON e.department_id = d.department_id;
"""
cursor.execute(query16)
df16 = pd.DataFrame(cursor.fetchall(), columns=[col[0] for col in cursor.description])
display(df16)
```

```python
# 17. List the employees along with their job titles and the location of their department.
query17 = """
SELECT e.first_name, e.last_name, j.job_title, l.city, l.state_province
FROM employees e
JOIN jobs j ON e.job_id = j.job_id
JOIN departments d ON e.department_id = d.department_id
JOIN locations l ON d.location_id = l.location_id;
"""
cursor.execute(query17)
df17 = pd.DataFrame(cursor.fetchall(), columns=[col[0] for col in cursor.description])
display(df17)
```

```python
query18 = """
SELECT d.department_name, COUNT(e.employee_id) AS employee_count
FROM departments d
LEFT JOIN employees e ON d.department_id = e.department_id
GROUP BY d.department_name;
"""
cursor.execute(query18)
df18 = pd.DataFrame(cursor.fetchall(), columns=[col[0] for col in cursor.description])
display(df18)
```

```python
query19 = """
SELECT e.first_name, e.last_name, e.job_id
FROM employees e
JOIN employees m ON e.manager_id = m.employee_id
WHERE e.job_id = m.job_id;
"""
cursor.execute(query19)
df19 = pd.DataFrame(cursor.fetchall(), columns=[col[0] for col in cursor.description])
display(df19)
```

```python
# 20. Display the names of employees who worked in different jobs in the past (using job_history).
query20 = """
SELECT DISTINCT e.first_name, e.last_name
FROM employees e
JOIN job_history jh ON e.employee_id = jh.employee_id;
"""
cursor.execute(query20)
df20 = pd.DataFrame(cursor.fetchall(), columns=[col[0] for col in cursor.description])
display(df20)
```
